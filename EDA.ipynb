# Univariate, Bivariate and MultiVariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Cars93.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
# pick only Model, Type, Price, Horsepower, EngineSize
df = df[['Model','Type','Price','Horsepower','EngineSize']]
```

```python
df.head()
```

```python
df.shape
```

## Univariate Analysis

```python
df['Type'].value_counts()
```

```python
df_small_type = df.loc[df['Type']=='Small']
df_midsize_type = df.loc[df['Type'] == 'Midsize']
df_large_type = df.loc[df['Type'] == 'Large']
```

```python
plt.plot(df_small_type['Horsepower'], np.zeros_like(df_small_type['Horsepower']),'x')
plt.plot(df_midsize_type['Horsepower'], np.zeros_like(df_midsize_type['Horsepower']),'o')
plt.plot(df_large_type['Horsepower'], np.zeros_like(df_large_type['Horsepower']),'*')
plt.xlabel('Horsepower')
plt.show()
```

```python
df_midsize_type.columns
```

```python
df_midsize_type[['Model','Type','Price','EngineSize']][df_midsize_type['Horsepower'] > 250]
```

## Bivairate Analysis

```python
sns.FacetGrid(df, hue="Type",height=5).map(plt.scatter, "Horsepower","EngineSize").add_legend()
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df,hue="Type",height=5)
```
